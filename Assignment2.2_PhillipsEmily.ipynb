```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#reading in data
#world-population excel spreadsheet
world_pop = pd.read_excel("ex2-2/world-population.xlsm")
```

```python
#looking at first five recordss
print(world_pop.head())
```

# Line Chart

```python
plt.plot(world_pop['Year'], world_pop['Population'])  # Plot the chart
plt.xlabel("Year")  # add X-axis label
plt.ylabel("Population")  # add Y-axis label
plt.title("Population Growth from 1960 to 2010")  # add title
plt.show()  # display
```

# Step Chart

```python
#matplotlib step function for step chart
plt.step(world_pop['Year'], world_pop['Population'])
#labels and title
plt.xlabel("Year")  # add X-axis label
plt.ylabel("Population")  # add Y-axis label
plt.title("Population Growth from 1960 to 2010")  # add title
plt.show()
```
